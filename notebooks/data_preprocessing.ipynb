This notebook demonstrates the preprocessing pipeline used to clean medical protocols and prepare them for the FAISS vector database.

```python
import json
import pandas as pd
from langchain.text_splitter import RecursiveCharacterTextSplitter

# 1. Load Raw Medical Data
with open('../data/medical_guidelines.json', 'r') as f:
    raw_data = json.load(f)

df = pd.DataFrame(raw_data)
print(f"Loaded {len(df)} professional guidelines.")
df.head()
```

```python
# 2. Demonstrate Text Splitting Logic
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

sample_text = f"Condition: {raw_data[0]['condition']}\nGuideline: {raw_data[0]['guideline']}"
chunks = text_splitter.split_text(sample_text)

print(f"Example splitting result: {len(chunks)} chunks created from sample.")
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ---\n{chunk}")
```

### Conclusion
The processed chunks are now ready to be embedded using `text-embedding-3-small` and stored in the FAISS index.
